 ### Filling or Replacing a missing value in a column with the "Mean"

```python
import pandas as pd
df=pd.read_csv("downloads\dirtydata.csv")
print(df)

x=df['Calories'].mean()
print(x)
df['Calories'].fillna(x,inplace=True)
print(df)
```

### Filling or Replacing a missing value in a column with the "Median"

```python
import pandas as pd
df=pd.read_csv("downloads\dirtydata.csv")
print(df)
x=df["Calories"].median()
print("Median : ",x)
df['Calories'].fillna(x,inplace=True)
print(df)
```

### ### Filling or Replacing a missing value in a column with the "Mode"

```python
import pandas as pd
df=pd.read_csv("downloads\dirtydata.csv")
print(df)
x=df["Calories"].mode()[0]
#remember to add [0] incase of 'mode'.
print("Mode : ",x)
df["Calories"].fillna(x,inplace=True)
print(df)
```

### Data Cleaning

```python
import pandas as pd
df=pd.read_csv("downloads\dirtydata.csv")
print(df.to_string())
x=df["Calories"].mean()
print("Mean : ",x)
df["Calories"].fillna(x,inplace=True)
print(df)
df["Date"]=pd.to_datetime(df["Date"])
print(df)
df.dropna(subset=["Date"],inplace=True)
print(df)
```

```python
print(df.loc[7,"Duration"])
df.loc[7,"Duration"] = 45
print(df)
```

### Setting Maxpulse = 130 when Maxpulse>130

```python
for x in df.index:
    if df.loc[x,"Maxpulse"]>130:
        df.loc[x,"Maxpulse"]=130
        
print(df)        
```

### Deleting Maxpulse data when Maxpulse>130

```python
for x in df.index:
    if df.loc[x,"Duration"]>130:
        df.drop(x,inplace=True)
print(df)    
```

### Removing Duplicates

```python
print(df.duplicated())
```

```python
df.drop_duplicates(inplace=True)
print(df)
```
