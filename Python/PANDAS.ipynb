## Pre - Concepts

- #### Understanding 'axis'

    A DataFrame object has two axes: **“axis 0”** and **“axis 1”:**

    **axis 0** : It is horizontal and represents rows. <br>
    **axis 1** : It is vertical and represents columns.
        
- #### Understanding 'inplace'

    When **inplace = False** , which is the default, then the operation is performed and it returns a copy of the object.<br> 
    You then need to save it to something.
    
    When **inplace = True** , the input dataframe is modified, which means it will return nothing and the dataframe is now updated.

### df.loc VS df.iloc

### df.loc
loc is label-based, which means that we have to specify the name of the rows and columns that we need to filter out.

**Note**: if the indices are not numbers, then we cannot slice our data frame. In that case, we need to use the iloc function.

```python
df.loc[0]    # Returns first row 
```

```python
df.loc[1]    # Returns 2nd row which is indexed as 1
```

```python
df.loc[1,'customerName']    # Returns the customername of the first row
```

```python
df.loc[0:5,'customerName']
```

```python
df.loc[1, 'customerNumber':'country']   

df.loc[df.creditLimit > 150000] # Returns all the details from customerNumber to country from the second row indexed as 0
```

```python
df.loc[0:5,'customerName':'country']
```

```python
df.loc[0:5, ['customerName', 'country', 'city']]
```

```python
df.loc[0:5]    # Returns rows from 0 to 5, includes 5th row also. Returns total 6 rows
```

```python
df.loc[:]    # Returns all the rows from start to end
```

```python
df.loc[df['creditLimit']>150000]    # Select only required columns with a condition

# df.loc[df.creditLimit > 150000] 

# df[df.creditLimit > 150000]

# df[df['creditLimit'] > 150000]    # All these returns same output
```

```python
df.loc[(df['creditLimit'] > 120000) & (df['country'] == 'USA')]
```

```python
df.loc[df.country == 'USA', ['customerName', 'country', 'city']].head()    # Returns specified columns with a filter applied
```

```python
df[df.country == 'USA'][['customerName', 'country', 'city']].head()    # Same as above
```

### iloc

iloc is integer index-based , we have to specify rows and columns by their **integer index**.<br>
i before the loc stands for indexing

```python
df.iloc[[0,1,2,3]]    # Passing a list of indexes
```

```python
df.iloc[0,1]    
```

```python
df.iloc[0][1]
```

```python
df.iloc[0:5, 1:5]    # includes 0 but excludes 5, same with columns includes 1 but excludes 5.
```
